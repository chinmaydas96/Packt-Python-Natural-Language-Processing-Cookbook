```python
pos_mapping = {'JJ':'a', 'JJR':'a', 'JJS':'a', 'NN':'n',
               'NNS':'n', 'VBD':'v', 'VBG':'v','VBN':'v',
               'VBP':'v', 'VBZ':'v'}
accepted_pos = {'a', 'v', 'n'}
```

```python
from pos_tagging import pos_tag_nltk, read_text_file
```

```python
def lemmatize_long_text(text):
    words = pos_tag_nltk(text)
    words = [(word_tuple[0], pos_mapping[word_tuple[1]] if \
      word_tuple[1] in pos_mapping.keys() else 
      word_tuple[1]) for word_tuple in words]

    words = [(lemmatizer.lemmatize(word_tuple[0]) if \
               word_tuple[1] in accepted_pos else \
               word_tuple[0],
             word_tuple[1]) for word_tuple in words]
    return words
```

```python
sherlock_holmes_text = read_text_file("sherlock_holmes_1.txt")
lem_words = lemmatize_long_text(sherlock_holmes_text)
print(lem_words)
```

## Removing stopwords

```python
import csv
import nltk
```

```python
csv_file = 'stopwords.csv'
with open(csv_file, 'r', encoding='utf-8') as f:
    reader = csv.reader(f, delimiter=',', quotechar='"')
    stopwords = [row[0] for row in reader]
```

```python
print(stopwords)
```

```python
import nltk
from nltk.probability import FreqDist
```

```python
with open('sherlock_holmes.txt', 'r', encoding='utf-8') as f:
    text = f.read()
text = text.replace('\n', ' ')
```

```python
words = nltk.tokenize.word_tokenize(text)
```

```python
freq_dist = FreqDist(word.lower() for word in words)
words_with_frequencies = [(word, freq_dist[word]) for word in freq_dist.keys()]
```

```python
words_with_frequencies
```

```python
sorted_words = sorted(words_with_frequencies, key=lambda tup: tup[1])
```

```python
stopwords = [tuple[0] for tuple in sorted_words if tuple[1] > 250]
stopwords
```

```python
length_cutoff = int(0.005 * len(sorted_words))
stopwords = [tuple[0] for tuple in sorted_words[-length_cutoff:]]
```
